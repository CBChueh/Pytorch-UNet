```python
import os
import numpy as np
from PIL import Image
folder = 'data/mask_jpg'
folder2 = 'data/masks'
os.makedirs(folder2, exist_ok=True)

for filename in os.listdir(folder):
    if filename.endswith(".jpg"):
        temp=np.asarray(Image.open(os.path.join(folder,filename)))
        temp=temp/255
        temp[temp<1]=0
        temp[temp>=1]=1
        np.save(os.path.join(folder2,filename.replace('jpg','npy').replace('mask','frame')), temp)
```
